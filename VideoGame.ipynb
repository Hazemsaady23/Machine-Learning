```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import LabelEncoder

from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import numpy as np
```

```python
df=pd.read_csv('vgsales.csv')
df.head(5)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.describe()
```

```python
df = df.dropna(subset=['Year','Publisher'])
```

```python
X = df[['Platform','Year','Publisher','Genre']].copy()
y = df['Global_Sales']
```

```python
for col in X.columns:
    X[col] = X[col].astype(str)
    le = LabelEncoder()
    X[col] = le.fit_transform(X[col])
```

```python

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
rf = RandomForestRegressor(n_estimators=400, random_state=42)
rf.fit(X_train, y_train)
```

```python
y_pred = rf.predict(X_test)

r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)

print("Random Forest Regression Results:")
print(f"R² Score: {r2:.3f}")
print(f"MAE: {mae:.3f}")
print(f"MSE: {mse:.3f}")
```

Linear Regression

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
df2 = pd.read_csv('vgsales.csv')
```

```python
df2.head(10)
```

```python
df2.describe()
```

```python
df2.isnull().sum()
```

```python
df2 = df2.dropna(subset=['Global_Sales','Year','Platform','Publisher','Genre'])

X = df2[['Platform','Year','Publisher','Genre']].copy()
y = df2['Global_Sales']

categorical_cols = ['Platform','Publisher','Genre']
numeric_cols = ['Year']
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_cols),
        ('num', StandardScaler(), numeric_cols)])
pipe = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('model', LinearRegression())
])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
pipe.fit(X_train, y_train)

y_pred = pipe.predict(X_test)
```

```python
r2 = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)

print("Linear Regression Results:")
print(f"R² Score: {r2:.3f}")
print(f"MAE: {mae:.3f}")
print(f"MSE: {mse:.3f}")
```
